# Read all csv files in a folder and extract basin id

```python
import os

folder = r"D:\Research\Projects\Hy2DL\data\CAMELS_DE\hourly" 

for fname in os.listdir(folder):
    if fname.endswith(".csv"):
        name_no_ext = os.path.splitext(fname)[0]
        id_part = name_no_ext.split("_")[-1]
        print(id_part)
```
